# 2. QAM Transmission over Fading Channels

## 2.1 Frequency-Nonselective Channels<a class="anchor" id="flatChannelModel"></a>

This Jupyter notebook considers an uncoded QAM transmission over fading channels with AWGN. The discrete-time model of a point-to-point link is illustrated in <a href="#ptp-link-discrete">Fig. 1</a>.

<figure><a name="ptp-link-discrete"></a>
    <center>
        <img src="Figures/part1_QAM_flatFading.png" align=center width=600/> <br><br>
        <figcaption>
            <b>Fig.1:</b> Discrete-time QAM transmission over flat fading channel
        </figcaption>
    </center>
</figure>

In contrast to a simple AWGN channel, the transmit signal $x[i]$ is first multiplied with a channel coefficient $h[i]$, before white Gaussian noise $w[i]$ is added. 

\begin{equation}
    y[i] = h[i] \cdot x[i] + w[i]
\end{equation}

In mobile radio communications, multipath propagation frequently occurs and the transmit signal $x[i]$ reaches the receiver over different paths. If the mutual delays between propagation paths are smaller than the symbol duration $\Ts$, the multipath components cannot be resolved by the receiver. In a diffuse scattering environment where the transmit signal impings uniformly from all directions at the receive antenna, the channel can be modeled by a single channel coefficient $h[i]$ being complex Gaussian distributed with zero mean and variance $\sigmaH=1$. This variance ensures that the channel is non-dissipative, i.e. the average transmit power $\sigmaX$ is identical to the average received signal power $\sigmaH \cdot \sigmaX$.

The multiplication with $h[i]$ leads to a random instantaneous received signal power $|h[i]|^2 \sigmaX$ which is chi-square distributed with two degrees of freedom. 

\begin{equation}
    p_{|\Hrv|^2}(\xi) = \frac{1}{\sigmaH} \cdot e^{ - \frac{\xi}{\sigmaH}}
\end{equation}

Therefore, the instantaneous SNR also becomes a random variable with the same distribution as $|\Hrv|^2$.

\begin{equation}
    \gamma[i] = |h[i]|^2 \cdot \frac{\sigmaX}{\sigmaW} 
\end{equation}



### 2.1.1 ML Detection for Frequency-Nonselective Channels<a class="anchor" id="MLdetectionflatChannel"></a>

For a coherent detection, the receiver needs to known the channel coefficients. Ideally, perfect channel state information (CSI) is available and the receiver knows all $h[i]$. Realistically, the receiver has to estimate the channel coefficients and only knows $\hat{h}[i] = h[i] + \epsilon[i]$ containing a certain estimation error $\epsilon[i]$. With the channel knowledge, the optimal maximum likelihood receiver can be realized.

\begin{equation}
    \hat{x}[i] = \argmin_{d \in \DD} | y[i] - h[i] d|^2  = \argmin_{d \in \DD} \left| \frac{y[i]}{h[i]} -  d \right|^2 \ .
\end{equation}

Hence, the ML decision rule already known from AWGN channels can be applied to the equalized signal $y_e[i] = \frac{y[i]}{h[i]} = x[i] + \frac{w[i]}{h[i]}$.


### 2.1.2 Error Probabilities for Frequency-Nonselective Channels<a class="anchor" id="ErrorProbFlatChannel"></a>

Due to the random characteristic of the SNR, the error probability

\begin{equation}
    \Ps[i] = \frac{1}{2} \cdot \mathrm{erfc} \left( \sqrt{ \left( \frac{\Delta_0}{2} \right)^2 |h[i]|^2 \cdot \frac{\sigmaX}{\sigmaW} } \right)
\end{equation}

becomes a function of $|h[i]|^2$ and time varying. For fast fading channels, the ergodic error probability is a meaningful performance measure. It is obtained by averaging $\Ps[i]$ w.r.t. the statistics of $|\Hrv|^2$. The exact solution requires a lengthy derivation. For 4-QAM, we approximately obtain

\begin{align}
    \bar{\Ps} 
    &= \Exp_{\Hrv} \left\{ \frac{1}{2} \cdot \mathrm{erfc} \left( \sqrt{ \left( \frac{\Delta_0}{2} \right)^2 |h|^2 \cdot \frac{\sigmaX}{\sigmaW} } \right) \right\}
    \approx 1 - \sqrt{ \frac{ \Es/N_0 }{ 2 + \Es/N_0 } } \ .
\end{align}

## 2.2 Frequency-Selective Channels<a class="anchor" id="frequencySelectiveChannelModel"></a>

If the delays of mulipath components exceed the symbol duration $\Ts$, intersymbol interference occurs. Now, the channel is not characterized by a single channel coefficient but an impulse response $h_{\mu}[i]$ of length $\Lh+1$. In the sequel, we assume that the channel is time invariant such that the time index $i$ can be dropped. Now, the received signal is obtained from a convolution of the transmit signal $x[i]$ with the channel impulse response leading to

\begin{equation}
    y[i] = (h \ast x)[i]+ w[i] = \sum_{\mu=0}^{\Lh} h_{\mu} \cdot x[i-\mu] + w[i] \ .
\end{equation}

The above equation describes a discrete time convolution which is corrupted by AWGN $w[i]$. The corresponding channel model is illustrated <a href="#frequency_selective_channel_model">Fig. 2</a>.

<figure><a name="frequency_selective_channel_model"></a>
    <center>
        <img src="Figures/part1_discrete_time_channel_model2.png" align=center width=600/> <br><br>
        <figcaption>
            <b>Fig.2:</b> Discrete-time model of frequency selective channel
        </figcaption>
    </center>
</figure>

Due to the channel memory, intersymbol interference (ISI) occurs and each sample $y[i]$ contains $\Lh+1$ data symbols $d[i-\mu]$ for $0 \leq \mu \leq \Lh$. Moreover, successive channel outputs $y[i]$ depend on each other and a symbol-wise detection  becomes highly suboptimal. Instead, a whole sequence has to be detected. A compact mathematical description is obtained with the vector notation. Stacking $K$ transmit symbols $d[i]$ into a vector $\xv$, and all $L$ received symbols $y[i]$ into $\yv = \begin{bmatrix} y[0] & \ldots & y[L-1] \end{bmatrix}^{\mathrm{T}}$, we obtain the system description

\begin{equation}
    \yv = \Hm \cdot \xv + \wv \ .
\end{equation}

The length $L$ depends on the number of transmit symbols. It usually amounts to $L=K+\Lh$ because the convolution increases the length of the input sequence. The channel matrix $\Hm$ is a Toeplitz matrix, i.e. it is a convolutional matrix with the structure

\begin{align*}
    \begin{bmatrix}
        y[0] \\ y[1] \\ \vdots \\ y[K+\Lh]  
    \end{bmatrix} 
    &= \begin{bmatrix}
            h_0	& 0	& \cdots & 0 \\
            h_1 & h_0 &	& \\ 
            \vdots & \vdots & \ddots & \\ 
            h_{\Lh} & & & h_0 \\
            & h_{\Lh} & & \\
            & & \ddots & \vdots \\
            0 & & & h_{\Lh}
        \end{bmatrix}
    \cdot
    \begin{bmatrix}
        x[0] \\ x[1] \\ \vdots \\ x[K]  
    \end{bmatrix} 
    + 
    \begin{bmatrix}
        w[0] \\ w[1] \\ \vdots \\ w[K+L_\mathrm{h}]  
    \end{bmatrix} \ .
\end{align*}

In many models, the channel coefficents are assumed to be complex Gaussian distributed with zero mean and variance $\sigma_{\Hrv,\mu}^2$. The average powers $\sigma_{\Hrv,\mu}^2$$\sigma_{\Hrv,\mu}^2$ of the channel coefficients $h_{\mu}$ are defined by the channel delay profile (PDP) describing the statistical distribution of the channel energy versus the delay $\mu$. Often, the average channel energy is assumed to be one, i.e. $\sum_{\mu=0}^{\Lh} \sigma_{\Hrv,\mu}^2=1$ holds. For a uniform PDP, $\sigma_{\Hrv,\mu}^2 = \frac{1}{\Lh+1}$ follows. Alternatively, exponential PDPs are frequently assumed leading to $\sigma_{\Hrv,\mu}^2 = A \cdot e^{- \alpha \mu}$ with $A = \left(\sum_{\mu=0}^{\Lh} e^{- \alpha \mu} \right)^{-1}$.

## 2.3 Concluding Remarks

* We have learned that frequency selective channels lead to intersymbol interference (ISI), which deteriorates the performance significantly. 
* The optimal detector would perform a maximum likelihood sequence detection by applying the Viterbi algorithm (see 2nd generation of cellulear mobile radio systems). 
* Unfortunately, the Viterbi algorithm does not scale with increasing bandwidth (more channel taps, i.e. longer channel impulse responses). hence, alternative approaches have to be found.
