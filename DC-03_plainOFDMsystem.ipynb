# 3. Orthogonal Frequency Division Multiplexing (OFDM)

## 3.1 Basic Principle of Multi-Carrier Transmission<a class="anchor" id="OFDMprinciple"></a>

From the previous exercise we have learned that frequency selective channels lead to intersymbol interference (ISI), which deteriorates the performance significantly. The optimal Viterbi detector performs a maximum likelihood sequence detection, but its complexity grows exponentially with the channel length $\Lh$. Hence, it becomes quickly infeasible for todays signal bandwidth.

            <b>Fig.1:</b> Principle of multi-carrier transmission
        </figcaption>

A completely different approach for a robust transmission over ISI channels is OFDM. The general idea is depicted in <a href="#MultiCarrier">Fig. 1</a>. The signal bandwidth $B$ over which the channel behaves frequency selective is divided into $N$ narrowband parts. The new subcarriers occupy a bandwidth $\frac{B}{N}$ in which the channel is approximately frequency non-selective. 


### 3.1.1 Discrete-time OFDM system model 

The complete discrete-time OFDM system model is given in <a href="#OFDMsystem">Fig. 2</a>. The transmitter first divides the bit stream into $N$ parallel streams. The bits in each of these streams are mapped onto data symbols taken from conventional PSK or QAM alphabets. In practical wireless communication systems, the same modulation scheme is used for all streams. Theoretically, they can be independently chosen and adapted to the individual channel conditions of a particular subcarrier. This requires appropriate channel knowledge at the transmitter which is hard to obtain in time-varying environments. 

The streams' data symbols are transmitted on orthogonal subcarriers to avoid intercarrier interference. This is efficiently implemented in a blockwise fashion using the inverse fast Fourier transform (IFFT). It transforms the data from the frequency domain into the time domain. In order to avoid ISI and to preserve the subcarriers' orthogonality, a guard interval has to be inserted in front of each IFFT output vector.

The receiver uses the guard interval for estimating a potential carrier frequency offsets (CFO) between transmitter and receiver. Afterwards the guard interval can be removed. Next, the data back is transformed back into the frequency domain using the FFT leading to $N$ parallel data streams. They are equalized in frequency domain by a simple scalar division with the corresponding channel coefficient, i.e. since each subcarrier is assumed to be frequency non-selective. The obtained estimates of the subcarriers' data symbols can now be detected and mapped onto the corresponding bit vectors. Finally, the bit streams are multiplexed into a single stream.

### 3.1.2. Guard Interval, Transmisssion and linear Equalizer
 
As shown in the <a href="#GuardInterval">Fig. 3</a>, a guard interval is inserted in front of each data block in the time domain. There are two reasons why the guard interval is essential for OFDM. First, successive OFDM symbols interfer in time domain due to the frequency selective behavior of the channel. This interference can be captured by the guard interval whose length $\Tg$ has to be longer than the channel impulse response $\Lh \Ts$. If this condition is fulfilled, removing the guard interval removes ISI as well. Second, the guard interval is not just a simple gap but contains a cyclic prefix of the current OFDM symbol. This ensures that the linear convolution of channel impulse response and transmit signal becomes a cyclic convolution (when looking at the samples of the core OFDM symbol). Applying the FFT at the receiver, the cylic convolution in the time domain is transformed into a scalar multiplication in the frequency domain. This means that the data symbols on the subcarriers are simply multiplied with the corresponding coefficients of the channel transfer function (being the discrete-time Fourier transform of the channel impulse response). Mathematically, we obtain

\begin{align}
    \yv[k] &= \Hm \cdot \xv[k] + \nv[k]
    \qquad \xrightarrow[\text{DFT}]{} \qquad
    \yv_{\mathrm f}[k] &= \Hm_{\mathrm f} \cdot \xv_{\mathrm f}[k] + \nv_{\mathrm f}[k] = \Hm_{\mathrm f} \cdot \dv[k] + \nv_{\mathrm f}[k]
\end{align}

with $\Hm$ beeing a cyclic Toeplitz matrix. These matrices are diagonalized by the Fourier transform. Hence, $\Hm_{\mathrm f}$ is a diagonal matrix having the channel transfer function on its diagonal. For subcarrier $i$, we obtain the scalar notation $Y_i[k] = H_i \cdot d_i[k] + N_i[k]$ in the frequency domain. This subcarrier-based scalar multiplication allows the simple equalization which itself is just a scalar division. 

\begin{align}
    \tilde{\dv}[k] 
        &= \Em_{\mathrm f} \cdot \yv_{\mathrm f}[k]
         = \Hm_{\mathrm f}^{-1} \cdot \yv_{\mathrm f}[k]
         = \dv[k] + \Hm_{\mathrm f}^{-1} \cdot \nv_{\mathrm f}[k] \\
    \tilde{d}_i[k] 
        &= E_{{\mathrm f},i} \cdot y_{{\mathrm f},i}[k]
         = H_i^{-1} \cdot y_{{\mathrm f},i}[k]
         = d_i[k] + H_{{\mathrm f},i}^{-1} \cdot n_{{\mathrm f},i}[k] 
\end{align}
 
The linear equalizer represents the least-squares (LS) solution and removes the channel influence completely (no factor remains in front of the data symbol $d_k[k]$). However, the white Gaussian noise is also multiplied with $H_i^{-1}$. For channels with deep fades, $H_i$ is small and, hence, $H_i^{-1}$ becomes very large. This leads to a noise amplification which is typical for LS estimators. Contrarily, the noise on strong subchannels is attenuated leading to high SNRs. Therefore, the performance of OFDM systems will be dominated by those subcarriers suffering from deep fades. A way out of this dilemma is the application of error control coding which is not part of this course.
