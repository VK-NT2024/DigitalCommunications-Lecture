# 7. Frequency Synchronization

In WiFi systems according to the IEEE 802.11 standard, no central clock is provided and multiple access is handled by the carrier sense multiple access (CSMA) technique. As a consequence,  the receiver needs to perform some processing steps to ensure time and frequency synchronization as well as channel estimation. Therefore, a specific frame structure is required containing not only data symbols (payload), but also training fields for the above purposes. The complete structure for the IEEE 802.11a standard is depicted in the <a href="#IEEE802.11_frame">Fig. 1</a>. It starts with a preamble consisting of the short training field (STF) and a long training field (LTF). 

* The short training field has the length of two OFDM symbols (8$\mus$) and contains ten periods of a sequence with duration 0.8$\mus$. Its autocorrelation functions consists of  multiple impulses with the largest peak at delay zero. The STF is used for time synchronization. 
* The long training field has the same length but contains pilot symbols for channel estimation. The two OFDM symbols of the LTF have a common guard interval whose length is twice as long as the regular guard interval of a single OFDM symbol. Therefore, channel impulse responses being longer than a regular guard interval can be correctly estimated. 

Next, a signal field informs the receiver about the type of coding and modulation as well as the size of the data field. This signal field and the subsequent service field are not used within the project. Finally, the data field contains the data symbols, i.e. the payload.

## 7.1 Influence of Carrier Frequency Offset

This notebook considers a carrier frequency offset (CFO) between transmitter and receiver in OFDM systems. The offset $\Delta f_0$ comes from a lack of synchronization between the local oscillator (LO) at the transmitter operating at a carrier frequency $f_0$ and the LO at the receiver operating at $\tilde{f}_0 = f_0 + \Delta f_0$. A CFO causes cumulative phase rotations in the time domain by which the orthogonality of sub-carriers is destroyed. For a distortionless and noise-free received signal, the CFO can be described by

\begin{equation*}
    y(t) = x(t) \cdot e^{j \Delta \omega_0 t} = x(t) \cdot e^{j 2 \pi \Delta f_0 t} 
\end{equation*}

Hence, estimating and compensating the CFO has to start already in the time domain before applying the FFT. The normalized CFO consists of three parts.

\begin{equation*}
    \varepsilon_\mathsf{CFO} = \varepsilon_\mathsf{IFO} + \varepsilon_\mathsf{FFO} + \varepsilon_\mathsf{RFO} = 2 \pi \Delta f_0 T_{\mathrm{MC}}
\end{equation*}

* Integer Frequency Offset (IFO): This offset is an integer multiple of the subcarrier spacing $\Delta F = T_{\mathrm{MC}}^{-1}$. Its compensation requires additional reference symbols (not addressed here).
* Fractional Frequency Offset (FFO): This offset is a fraction of the subcarrier spacing and is compensated in time domain.
* Residual Frequency Offset (RFO): $\varepsilon_\mathsf{RFO} = \varepsilon_\mathsf{CFO} - \widehat{\varepsilon}_\mathsf{IFO} - \widehat{\varepsilon}_\mathsf{FFO} $ <br>
 This is the remaining offset after IFO and CFO compensation. It may be tolerated or estimated and compensated by refined algorithms.


## 7.2 Estimation of Carrier Frequency Offset with STF

Since the CFO leads to inter-carrier interference in the frequency domain, it has to be compensated in the time domain before the FFT is computed at the receiver. In the following explanations, we assume that the preamble detection has been successfully performed and the beginning of a frame is known. In a first step, no multi-path fading occurs and only AWGN shall disturb the transmission.

Having a reference signal $x_{\mathrm{ref}}[\ell]$ known to the receiver, a direct way to estimate the CFO is the sample-wise multiplication of the received sequence $y[\ell]$ with $x_{\mathrm{ref}}^*[\ell]$. Without CFO, the product equals the squard magnitudes plus noise. 

\begin{equation*} 
    y[\ell] \cdot x_{\mathrm{ref}}^*[\ell] = |x_{\mathrm{ref}}[\ell]|^2 + \tilde{n}[\ell]
\end{equation*}

With CFO, phase rotations occur and can be estimated.

\begin{equation*} 
    y[\ell] \cdot x_{\mathrm{ref}}^*[\ell] = |x_{\mathrm{ref}}[\ell]|^2 \cdot e^{j 2 \pi \Delta f_0 \ell \Ts} + \tilde{n}[\ell]
\end{equation*}

The CFO produces the same phase difference between adjacent symbols. Hence, averaging the phase differences between successive samples delivers

\begin{equation*} 
    \bar{\varphi} = \frac{1}{L-1} \sum_{\ell=0}^{L-2} \left( \arg \left\{ y[\ell+1] \cdot x_{\mathrm{ref}}^*[\ell+1] \right\} - \arg \left\{ y[\ell] \cdot x_{\mathrm{ref}}^*[\ell] \right\} \right) \; .
\end{equation*}

With $\varphi = \Delta f_0 \Ts$, the CFO can now be estimated. The CFO normalized to the sampling rate becomes

\begin{align*}
    \widehat{\varepsilon}_{\mathsf{CFO} }
        &= \Delta f_0 \Ts 
         = - \frac{\arg \{ \bar{\varphi} \}}{2 \pi} \; .
\end{align*}


In WiFi systems according to the IEEE 802.11 standard, the short training field might be used for the CFO estimation. The LTF will work as well. However, in cellular systems like 4G and 5G systems, no preamble is available for synchronisation purposes. For these OFDM based systems, the following algorithm is suitable.



## 7.3 Estimation of Carrier Frequency Offset with Schmidl-Cox Algorithm

OFDM systems enable another way for CFO estimation. The cyclic prefix (CP) of individual OFDM symbols can be compared with the tails of the core symbols. This approach is known as the Schmidl-Cox algorithm and provides a first (rough) estimate of the CFO. The principal steps are similar to the ones described above, but the cyclic prefix now serves as the reference signal. Particularly, it has to be considered that the CP is separated by $N$ samples from the tail (FFT size).
 
* First, the CP of OFDM symbol $k$ is correlated with its tail. This is equivalent to the multiplication with the reference signal and averaging. The phase estimate becomes

\begin{equation*} 
    \varphi[k] = \arg \left\{ \sum_{\mu=-\NG}^{-1} y_\mu [k] y^*_{\mu+N}[k] \right\}
\end{equation*}

* Repeating this correlation over multiple OFDM symbols and summing the results provides stronger noise suppression and yields

\begin{equation*} 
    \bar{\varphi} = \arg \left\{ \sum_k \sum_{\mu=-\NG}^{-1} y_\mu [k] y^*_{\mu+N}[k] \right\}
\end{equation*}

* If no FFO disturbs the transmission, the phases of CPs and tails are almost identical and the correlation represents an estimate of the CP's energy and is approximately real valued. In case of a non-zero CFO, a phase rotation $e^{j \Delta \omega_0 N \Ts}$ between corresponding samples $y_\mu[k]$ and $y_{\mu+N}[k]$ in CP and tail occurs. It can be estimated by

\begin{align*}
    \widehat{\varepsilon}_{\mathsf{CFO} }
        &= \Delta f_0 \frac{T_{\mathrm{MC}}}{N} = \Delta f_0 \Ts 
         = - \frac{\arg \{ \bar{\varphi} \}}{2 \pi}
\end{align*}


## 7.4 Compensation of Carrier Frequency Offset

The phase rotation introduced by the CFO can be compensated in the time domain.

\begin{equation*}
    \widetilde{y}[\ell] = y[\ell] \cdot e^{-j 2 \pi \ell \widehat{\varepsilon}_{\mathsf{FFO}}}
\end{equation*}

Hence, the presence of the cyclic prefix in OFDM systems does not only provide robustness against ISI, but enables CFO estimation.
