# 5. Channel Estimation

In WiFi systems according to the IEEE 802.11 standard, no central clock is provided and multiple access is handled by the carrier sense multiple access (CSMA) technique. As a consequence,  the receiver needs to perform some processing steps to ensure time and frequency synchronization as well as channel estimation. Therefore, a specific frame structure is required containing not only data symbols (payload), but also training fields for the above purposes. The complete structure for the IEEE 802.11a standard is depicted in the <a href="#IEEE802.11_frame">Fig. 1</a>. It starts with a preamble consisting of the short training field (STF) and a long training field (LTF). The short training field has the length of two OFDM symbols (8$\mus$) and contains ten periods of a sequence with duration 0.8$\mus$. Its autocorrelation functions consists of  multiple impulses with the largest peak at delay zero. The STF is used for time synchronization. The long training field has the same length but contains pilot symbols for channel estimation. The two OFDM symbols of the LTF have a common guard interval whose length is twice as long as the regular guard interval of a single OFDM symbol. Therefore, channel impulse responses being longer than a regular guard interval can be correctly estimated. Next, a signal field informs the receiver about the type of coding and modulation as well as the size of the data field. This signal field and the subsequent service field are not used within the project. Finally, the data field contains the data symbols, i.e. the payload.

<figure><a name="IEEE802.11_frame"></a>
    <center>
        <img src="Figures/part3_80211a_packet_structure2.png" align=center width=600/><br>
        <figcaption>
            <b>Fig.1:</b> Frame structure of IEEE 802.11a standard
        </figcaption>
    </center>
</figure>

The time-frequency grid is depicted in the <a href="#IEEE802.11_time_frequency">Fig. 2</a>. The signal of the short training field is generated by using only a subset of subcarriers with indices $[ \pm 4, \pm 8, \pm 12, \pm 16, \pm 20, \pm 24]$. This explains the periodicity in the time domain. In the data field, four pilot tones are included at subcarriers $[\pm 7, \pm 21]$ for fine synchronization. They are used for estimating the residual carrier frequency offset (CFO) in the frequency domain (see experiment DC-07). Over these subcarriers, no data symbols are transmitted.
