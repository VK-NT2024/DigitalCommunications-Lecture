This Jupyter notebook considers an uncoded QAM transmission over an AWGN channel. The simplified continuous-time model of the point-to-point link is illustrated in <a href="#ptp-link-cont">Fig. 1</a>.

        </figcaption>
    </center>
</figure>

The QAM (Quadrature Amplitude Modulation) mapper maps $m$ information bits $u[k]$ onto a complex-valued QAM symbols $d[i] \in \DD$. The alphabet $\DD$ contains all $M=2^m$ QAM symbols. The symbol $d[i]$ are passed to the transmit filter $g_{\mathrm{T}_\mathrm{x}}(t)$ to obtain a physical signal 

\begin{equation}
    x(t) = \Ts \cdot \sum_i d[i] \cdot g_{\mathrm{T}_{\mathrm{x}}}(t-i \Ts)
\end{equation}

being continuous in time and amplitude. The transmission is corrupted by additive white Gaussian noise $w(t)$ delivering $y(t) = x(t) + w(t)$ at the receiver. Here, the receive filter $g_{\mathrm{R}_\mathrm{x}}(t)$ performs lowpass filtering followed by sampling at symbol rate $\Ts^{-1}$. 

\begin{align}
    y[i] &= \left. \left( g_{\mathrm{R}_\mathrm{x}}(t) \ast y(t) \right) \right|_{t = i \Ts}
          = \Ts \cdot \sum_l d[l] \cdot g_{\mathrm{R}_{\mathrm{x}}}(t) \ast g_{\mathrm{T}_{\mathrm{x}}}(t-l \Ts) \big|_{t = i \Ts} 
          + g_{\mathrm{R}_{\mathrm{x}}}(t) \ast w(t) \big|_{t = i \Ts} \\
        &= \Ts \cdot \sum_l d[l] \cdot g(t-l \Ts) \big|_{t = i \Ts} + w[i]
\end{align}

The term $w[i]$ represents the filtered noise sampled at $t=i \Ts$. It remains white and Gaussian distributed the receive filter fulfills certain conditions, e.g. its squared transfer function has a symmetric Nyquist slope. In order to maximize the signal to noise ratio (SNR), the receive filter is matched to the transmit filter, i.e. $g_{\mathrm{R}_\mathrm{x}}(t)=g_{\mathrm{T}_\mathrm{x}}(-t)$ holds. Transmit and receive filter are usually designed such that the overall impulse response $g(t) = g_{\mathrm{R}_\mathrm{x}}(t) \ast g_{\mathrm{T}_\mathrm{x}}(t)$ fulfills the first Nyquist criterion, i.e.

\begin{equation}
    g(i T_\mathrm{s}) = 
    \begin{cases}
        \Ts^{-1} & \text{for } i=0 \\
        0 & \text{else} \; .
    \end{cases}
\end{equation}

If sampling is performed at ideal times $i \Ts$, successive symbols do not interfere with each other (no intersymbol interference, ISI) and one received symbol $y[i]$ only depends on the associated data symbol $d[i]$. 

\begin{equation}
    y[i] = \Ts \cdot \sum_l d[l] \cdot g(l \Ts - i \Ts) + w[i] = d[i] + w[i] 
\end{equation}

and the discrete-time system model shown in <a href="#ptp-link-discrete">Fig. 2</a>.

## 1.2 Maximum-Likelihood (ML) Detection 

The optimal Maximum-Likelihood detector works on a symbol-by-symbol basis because no ISI affects the transmission and the noise is white. Generally, it maximizes the likelihood function and yields the estimate

\begin{equation}
    \hat{d}^{(\mathsf{ML})}[i] 
        = \arg \! \max_{\tilde{d} \in \DD} p(y[i] \mid \tilde{d})
        = \arg \! \max_{\tilde{d} \in \DD} \log p(y[i] \mid \tilde{d}) \; .
\end{equation}

For Gaussian noise, it determines that symbol $\hat{d}[i] \in \DD$ which has the smallest Euclidean distance to the received symbol $y[i]$. In this case, we obtain

\begin{equation}
    \hat{d}^{(\mathsf{ML})}[i] = \arg \! \min_{\tilde{d} \in \DD} |y[i] - \tilde{d}|^2 \; .
\end{equation}

An important parameter of the AWGN channel is the signal-to-noise ratio (SNR). Using the power spectral density $N_0$ of the noise and the energy $\Es$ per symbol, it is defined to 

\begin{equation}
    \gamma = 
        \begin{cases}
            \frac{\Es}{N_0} & \text{for complex-valued alphabets} \\ 
            \frac{\Es}{N_0/2} & \text{for real-valued alphabets} 
    \end{cases}
\end{equation}


### 1.2.1 Analytical Error Probabilities

The performance of digital communication systems is usually measured by error probabilities. For the uncoded transmission, we distinguish bit and symbol error probabilities. A detection error occurs if the detector decides in favor to a wrong symbol, i.e. $\hat{D} \neq D$. This happens for Gaussian noise and ML detection when the competing symbol $\tilde{d} \neq d$ is closer to $y$ than the true symbol $d$. The probability for this event is called pairwise error probability and depends on the SNR and the modulation scheme. Its analytical computation requires the integration of the likelihood function over a particular interval. Unfortunately, the Gaussian distribution cannot be integrated in closed form and numerical solutions have to be used. The are represented by the Gaussian error function $\mathrm{erf}(x)$ and its complement $\mathrm{erfc}(x)=1-\mathrm{erf}(x)$. They are defined by 

\begin{align*}
    \mathrm{erf}(x) &= \frac{2}{\sqrt{\pi}} \cdot \int_0^x e^{- \xi^2} \ d\xi \\
    \mathrm{erfc}(x) &= 1 - \mathrm{erf}(x)
    = \frac{2}{\sqrt{\pi}} \cdot \int_x^{\infty} e^{- \xi^2} \ d\xi .
\end{align*}

The symbol error probabilities for different linear modulation schemes depend on the SNR and the minimum squared Euclidean distance $\Delta_0^2$ of the signal alphabet. The latter is usually normalized on the average transmit power $\sigmaD$.

\begin{equation}
    \Delta_0^2 = \min_{D,\tilde{D} \in \DD, D \neq \tilde{D}} 
        \frac{|D - \tilde{D}|^2}{\sigmaD}
\end{equation}

We obtain the following expressions for ASK, PSK and QAM.

\begin{align*}
    \Ps^{M-\mathsf{ASK}} 
    &= \frac{M-1}{M} \cdot \mathrm{erfc} \left( \sqrt{ \frac{3}{M^2-1} \frac{\Es}{N_0} } \right)
     = \frac{M-1}{M} \cdot \mathrm{erfc} \left( \sqrt{ \left( \frac{\Delta_0}{2} \right)^2 \frac{\Es}{N_0} } \right) \\
    \Ps^{M-\mathsf{QAM}} 
    &\approx 2 \frac{\sqrt{M}-1}{\sqrt{M}} \cdot \mathrm{erfc} \left( \sqrt{ \frac{3}{2(M-1)} \frac{\Es}{N_0} } \right) 
    = 2 \frac{\sqrt{M}-1}{\sqrt{M}} \cdot \mathrm{erfc} \left( \sqrt{ \left( \frac{\Delta_0}{2} \right)^2 \frac{\Es}{N_0} } \right) \\
    \Ps^{M-\mathsf{PSK}} 
    &\approx \mathrm{erfc} \left( \sqrt{ \sin^2 \left( \frac{\pi}{M} \right) \frac{\Es}{N_0} } \right)
     = \mathrm{erfc} \left( \sqrt{ \left( \frac{\Delta_0}{2} \right)^2 \frac{\Es}{N_0} } \right)
\end{align*}


### 1.2.2 Detemining Error Rates by Monte-Carlo Simulations

In a Monte Carlo simulation, they can be determined by comparing the transmitted bits/symbols with the detected ones. Counting the number of different bits/symbols and dividing by the total number of transmitted bits/symbols provides the corresponding error rates. With the indicator function 

\begin{equation}
    \delta(x) =  
        \begin{cases} 
            1 & \text{for } x \neq 0 \\ 
            0 & \text{for } x=0 
        \end{cases}
\end{equation}

the error rates can be expressed by

\begin{align}
    \Ps &= \frac{ 1 }{ \Nx } \cdot \sum_{i=0}^{\Nx-1} \delta(d[i] - \hat{d}[i]) \\
    \Pb &= \frac{ 1 }{ \Nu } \cdot \sum_{k=0}^{\Nu-1} \delta(u[k] - \hat{u}[k]) \approx \frac{\Ps}{m} \ .
\end{align}

$\Nx$ and $\Nu$ denote the number of data symbols and information bits, respectively. Certainly, $\Ps$ and $\Pb$ are identical for binary modulation schemes.
