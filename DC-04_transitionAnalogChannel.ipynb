This notebook focuses on the transition from the digital to the analog domain. Therefore, we will have a closer look on the transmit filter. It has to be designed such that the spectral repetitions are sufficiently suppressed. Usually, root raised cosine filter with impulse response 

\begin{equation}
    g_{\mathrm{RRC}}(t) = \frac{4 r \frac{t}{\Ts} \cos \left(\pi (1+r) \frac{t}{\Ts} \right) + \sin \left( \pi (1-r) \frac{t}{\Ts} \right)}{\left( 1- \left(4r \frac{t}{\Ts} \right)^2 \right) \pi t}
\end{equation}

and the transfer function 

\begin{equation}
    G_{\mathrm{RRC}}(j \omega) = 
    \begin{cases} 
        1 & \text{für } \frac{|\omega|}{\omega_{\mathrm{N}}} \leq 1-r \\
        \cos \left( \frac{\pi}{4r} \left( \frac{|\omega|}{\omega_{\mathrm{N}}} - (1-r) \right) \right)
        & \text{für } 1 - r \leq \frac{|\omega|}{\omega_{\mathrm{N}}} \leq 1+r \\
        0 & \text{für } \frac{|\omega|}{\omega_{\mathrm{N}}} \geq 1+r 
    \end{cases}
\end{equation}

are employed. They are parametrized by the roll-off factor $r \in [0,1]$. For $r=0$, the ideal lowpass filter is obtained, for $r=1$, the transfer function has a cosine shape. The Nyquist rate $\fN = \frac{1}{2 \Ts}$ relates the RF bandwidth $B = 2 (1+r) \fN$ to the symbol rate $\Ts^{-1}$. The corresponding angular frequency is $\omega_{\mathrm{N}} = 2 \pi \fN$. If the receiver employes the same root raised cosine filters (matched filter), the overall characteristic of transmit and receive filter becomes a raised cosine characteristic with impulse response

\begin{equation}
    g_{\mathrm{RC}}(t) = \frac{1}{\Ts} \cdot \frac{\sin(\pi t/\Ts)}{\pi t/\Ts} \cdot \frac{\cos(r \pi t/\Ts)}{1-(2rt/\Ts)^2}
\end{equation}

and transfer function

\begin{equation}
    G_{\mathrm{RC}}(j \omega) = 
    \begin{cases} 
        1 & \text{für } \frac{|\omega|}{\omega_{\mathrm{N}}} \leq 1-r \\
        \frac{1}{2} \cdot \left[ 1 + \cos \left( \frac{\pi}{2r} \left( \frac{|\omega|}{\omega_{\mathrm{N}}} - (1-r) \right) \right) \right]
        & \text{für } 1 - r \leq \frac{|\omega|}{\omega_{\mathrm{N}}} \leq 1+r \\
        0 & \text{für } \frac{|\omega|}{\omega_{\mathrm{N}}} \geq 1+r  \; .
    \end{cases}
\end{equation}

Due to finite length approximation of the filter impulse responses, this does not suffice to meet the spectral mask constraints. Therefore, idle subcarriers at the band edges are introduced.

## 4.2 Influence of Nonlinear Distortions (Power Amplifier)

In the equivalent baseband, there exist a complex-valued model of the power amplifier shown in the <a href="#power_amplifier">Fig. 4</a>. It can be split into two parts modelling the input-output behavior with respect to the magnitude and the phase. Regarding the amplitudes, real and imaginary parts can be treated independently. Here, a simple clipping at a maximum threshold or more sophisticated functions like Class A or Class AB amplifiers can be applied. The phase model has the magnitude $|x_{\mathrm{in}}|$ as input. For simple clipping, the phase is not changed at all, while for Class A and Class AB amplifiers, highly nonlinear phase distortions occur. The most severe effect of nonlinear disortions is the resultig out-of-band radiation. The output signals of nonlinear systems contain all linear combinations of the spectral components at their inputs. For a multi-carrier system with many input frequencies, frequency components far outside of the desired 20 MHz band can occur and the spectral mask is heavily violated.

One way to circumvent nonlinear distortion at least to a certain extent is the introduction of an input backoff. Here, the power of the input signal is reduced such that the peak powers mostly stay within the linear range of the power amplifier. The input backoff 

\begin{equation*}
    \mathrm{IBO}_{\mathrm{dB}} 
    = 20 \log_{10} \frac{ \max{| x_{\mathrm{out}}(t) |} }{ \sqrt{ \mathrm{E} \left \{ | x_{\mathrm{in}}(t)|^2 \right \}} }
    = 10 \log_{10} \frac{ \max{| x_{\mathrm{out}}(t) |^2} }{ \mathrm{E} \left \{ | x_{\mathrm{in}}(t)|^2 \right \} }
\end{equation*}

is defined as the ratio between the maximum peak of the squared magnitude at the amplifier output and its average input power. <a href="#nonlinear1">Fig. 5</a> illustrate this effect for simple clipping. The two graphs on the left show the simple clipping characteristic of the amplifier and the probability density funtion of the input magnitude (Rayleigh distribution). The graphs in the middle illustrate the effect of no or small input backoff. Here, clipping affects most of the input amplitudes and occurs with a large probability. The graphs on the right illustrate the case of a large input backoff. Here, only a small fraction of amplitudes is affected by clipping and the corresponding probability is significantly reduced.
